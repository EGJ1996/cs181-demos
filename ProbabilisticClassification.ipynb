```python
data_neg = [l.strip().split() for l in open("rt-polarity.neg", encoding = "ISO-8859-1")]
data_pos = [l.strip().split() for l in open("rt-polarity.pos", encoding = "ISO-8859-1")]
dictionary = {}
rev_dictionary = {}
counts = Counter()
for sent in data_pos + data_neg:
    for word in sent:
        counts[word] += 1
        if word not in dictionary:
            key = len(dictionary)
            dictionary[word] = key
            rev_dictionary[key] = word
```

```python
npos = len(data_pos)
nneg = len(data_neg)
n = npos + nneg
m = len(dictionary)
X = np.zeros((n, m))
y = np.zeros((n,))
y[:npos] = 1
```

```python
for i,sent in enumerate(data_pos + data_neg):
    for word in sent:
        j = dictionary[word]
        X[i, j] += 1
```

```python
def encode(sent):
    x = np.zeros((1, m))
    for word in sent.split():
        if word in dictionary:
            x[0, dictionary[word]] += 1
    return x
```

```python
model = sklearn.naive_bayes.MultinomialNB()
```

```python
X.shape, y.shape
```

```python
model.fit(X, y)
```

```python
features = []
for word, _ in counts.most_common(1000):
    j = dictionary[word]
    pi = model.feature_log_prob_[:, j]
    print("%10s \t %d \t %s"%(word, j, pi))
    features.append((word, pi[1]/pi[0]))
```

```python
features.sort(key=lambda x: x[1])
for w, s in features[:10]:
    print(w, s)
```

```python
for w, s in features[-10:]:
    print(w, s)
```
